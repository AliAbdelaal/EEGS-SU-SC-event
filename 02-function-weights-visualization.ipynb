# How weights would change the function shape

Let's take a quick look at how giving a weight to a function could change it's shape and eventually making it fit the data.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import animation
from IPython.display import HTML
```

# Linear function

$$ Y = X $$

```python
plt.rcParams['figure.figsize'] = (10, 10)
x = np.arange(-5, 5, .1)
y = x
plt.plot(x, y, '.')
plt.grid()
```

```python
fig, ax = plt.subplots()
ax.set_xlim(( -5, 5))
ax.set_ylim((-50, 50))
ax.grid()

line, = ax.plot([], [], label=r'$newX$')
orig_plot, = ax.plot([], [], label=r'$x$')
text = ax.text(3.5, 35, "weight = 1")

ax.legend()

def init():
    line.set_data([], [])
    orig_plot.set_data([], [])
    return (line,)

def animate(i):
    new_y = i*y
    line.set_data(x, new_y)
    orig_plot.set_data(x, y)
    text.set_text(f"a = {i}")
    return (line, orig_plot)

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=np.arange(-100, 100), interval=50, blit=True)
HTML(anim.to_html5_video())
```

```python
fig, ax = plt.subplots()
ax.set_xlim(( -5, 5))
ax.set_ylim((-50, 50))
ax.grid()

line, = ax.plot([], [], label=r'$newX$')
orig_plot, = ax.plot([], [], label=r'$x$')
text = ax.text(3.5, 35, "weight = 1")

ax.legend()

def init():
    line.set_data([], [])
    orig_plot.set_data([], [])
    return (line,)

def animate(i):
    new_y = i+y
    line.set_data(x, new_y)
    orig_plot.set_data(x, y)
    text.set_text(f"b = {i}")
    return (line, orig_plot)

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=np.arange(-50, 50), interval=50, blit=True)
HTML(anim.to_html5_video())
```
